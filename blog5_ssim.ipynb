# OpenCV Image Classifier

```python
from skimage.measure import compare_ssim as ssim 
#best method to compare images. Measures structural similarities index between two images
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
#to calculate Mean Squared Error
import cv2
#To load imaged. From module OpenCV
```

```python
def mse(imageA, imageB):
    err= np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err
```

```python
def compare_image(imageA, imageB, title):
    m= mse(imageA, imageB)
    s= ssim(imageA, imageB)
    fig= plt.figure(title)
    plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))    
    
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis("off")
    # show the second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageB, cmap = plt.cm.gray)
    plt.axis("off")
    
    plt.show()
```

```python
#Loading the images 
dog_1= cv2.imread('images/dog-1.png')
dog_2= cv2.imread('images/dog-2.png')
dog_3= cv2.imread('images/dog-3.png')
gate= cv2.imread('images/jp_gates_original.png')
cat_1= cv2.imread('images/cat-1.png')
cat_2= cv2.imread('images/cat-2.png')
cat_3= cv2.imread('images/cat-3.png')
```

```python
#Resizing the images look into zero patting
dog_1= cv2.resize(dog_1, (10000, 10000)) 
dog_2= cv2.resize(dog_2, (10000, 10000)) 
dog_3= cv2.resize(dog_3, (10000, 10000)) 
gate= cv2.resize(gate,(10000, 10000))
cat_1= cv2.resize(cat_1, (10000, 10000))
cat_2= cv2.resize(cat_2, (10000, 10000))
cat_3= cv2.resize(cat_3, (10000, 10000))
```

```python
#Converting to grayscale cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
dog_1= cv2.cvtColor(dog_1, cv2.COLOR_BGR2GRAY)
dog_2= cv2.cvtColor(dog_2, cv2.COLOR_BGR2GRAY)
dog_3= cv2.cvtColor(dog_3, cv2.COLOR_BGR2GRAY)
gate= cv2.cvtColor(gate, cv2.COLOR_BGR2GRAY)
cat_1= cv2.cvtColor(cat_1, cv2.COLOR_BGR2GRAY)
cat_2= cv2.cvtColor(cat_2, cv2.COLOR_BGR2GRAY)
cat_3= cv2.cvtColor(cat_3, cv2.COLOR_BGR2GRAY)
```

```python
# initialize the figure
fig = plt.figure("DOGS a gate and CATS!")
images = ("First Dog", dog_1), ("Second Dog", dog_2), ("Third Dog", dog_3),
        ('Gate', gate), 
        ('First Cat', cat_1), ('Second Cat', cat_2), ('Third Cat', cat_3)
# loop over the images
for (i, (name, image)) in enumerate(images):
    # show the image
    ax = fig.add_subplot(1, 7, i + 1)
#     ax.set_title(name)
    plt.imshow(image, cmap = plt.cm.gray)
    plt.axis("off")

# show the figure
plt.show();
```

```python
compare_image(dog_1, dog_1, 'same dogs')
```

```python
compare_image(cat_1, cat_1, 'same cats')
```

```python
# compare the images
compare_image(dog_1, dog_2, "First Dog vs. Second Dog")
compare_image(dog_1, dog_3, "First Dog vs. Third Dog")
compare_image(dog_2,dog_3, "Second Dog vs. Third Dog")
```

```python
# compare the images
compare_image(cat_1, cat_2, "First cat vs. Second cat")
compare_image(cat_1, cat_3, "First cat vs. Third cat")
compare_image(cat_2,cat_3, "Second cat vs. Third cat")
```

```python
# compare_image(dog_1, gate, 'First Dog vs. Gate')
# compare_image(dog_2, gate, 'Second Dog vs. Gate')
compare_image(dog_3, gate, 'Third Dog vs. Gate')
```

```python
compare_image(dog_1, cat_1, "First Dog vs. First Cat")
compare_image(dog_1, gate, 'First Dog vs. Gate')
compare_image(cat_1, gate, 'First cat vs. Gate')
```
